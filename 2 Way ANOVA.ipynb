```python
import pandas as pd
innerjoin = pd.read_csv('/Users/nuthakki/Documents/GA work/Judy - RadAI/innerjoin(FINAL TABLE TO USE).csv')
innerjoin
```

```python
#Pivoting doctor of machine_accuracy values
m = innerjoin.pivot(columns='doctor', values='machine_accuracy')
m1= pd.DataFrame(data=m)
m1.columns = ['a1', 'b1', 'c1', 'd1']
m1
```

```python
#Pivoting doctor of human_accuracy values
h = innerjoin.pivot(columns='doctor', values='human_accuracy')
h1= pd.DataFrame(data=h)
h1
```

```python
#Concatenating both the pivot tables of doctor
dfs = [h, m]
test = pd.concat(dfs, axis=1)
test
```

```python
#Replacing Nan with zeroes
import numpy as np
test[np.isnan(test)] = 0
test
```

```python
#Two-way ANOVA between human_accuracy of doctors (a, b, c, d') with all categories (a, b, c, d') of doctors from machine_accuracy 
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.tools import FigureFactory as FF

import numpy as np
import pandas as pd
import scipy

import statsmodels
import statsmodels.api as sm
from statsmodels.formula.api import ols
writer = pd.ExcelWriter('/Users/nuthakki/Desktop/anova_table_doctor.xlsx', engine='xlsxwriter')
no = 1
#pd.set_option('display.float_format', lambda x: '%.3f' % x)  #Suppress scientific notation

x = [test['a'], test['b'], test['c'], test['d']]

for i in x:
    formula = "i ~ C(a1) + C(b1) + C(c1) + C(d1)"
    model = ols(formula, data=test).fit()
    aov_table = statsmodels.stats.anova.anova_lm(model, typ=2)
    print(aov_table)
    aov_table.to_excel(writer, sheet_name='sheet1', startrow=int(no))
    no = no+8
    print(no)
writer.save()
#aov_table.to_csv('/Users/nuthakki/Desktop/anova_table1.csv')
        
```
